```python
from pathlib import Path
import tiktoken
import anthropic
import pandas as pd
import requests
from dotenv import load_dotenv
from anthropic import Anthropic
import os
load_dotenv()
os.environ["ANTHROPIC_API_KEY"] = os.getenv("ANTHROPIC_API_KEY")
from statistics import mean

def get_tokens_gpt(text):
    encoding = tiktoken.get_encoding("o200k_base")
    tokens = encoding.encode(text)
    return len(tokens)
    

def get_tokes_claude(text):
    client = anthropic.Anthropic()


    response = client.messages.count_tokens(
    model="claude-opus-4-6",
    messages=[{"role": "user", "content": text}],
)

    tokens = response.input_tokens
    return tokens


def load_calls_as_strings(folder="calls_german"):
    folder = Path(folder)
    gpt = []
    claude = []

    for fp in sorted(folder.glob("*.md")):
        text = fp.read_text(encoding="utf-8").strip()
        gpt.append(get_tokens_gpt(text))
        claude.append(get_tokes_claude(text))

    return gpt, claude
```

```python
gpt_tokens, claude_tokens = load_calls_as_strings()
```
